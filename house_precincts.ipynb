```python
from fuzzywuzzy import fuzz
```

```python
mit_precinct_list = list(mit_precinct_names)
mit_precinct_list
```

```python
match_dict = {}
for name in precinct_names:
    ratios = list(map(lambda x: fuzz.ratio(name, x), mit_precinct_list))
    f = lambda i: ratios[i]
    argmax_index = max(range(len(mit_precinct_list)), key=f)
    #print(argmax_index)
    match_dict[name] = mit_precinct_list[argmax_index]

match_dict
```

```python
print(len(match_dict.values()), len(set(match_dict.values())))
```
